```python
import pandas as pd
```

```python
import numpy as np
```

```python
df = pd.read_csv(r'http://github.com/YBI-Foundation/Dataset/raw/main/Boston.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.columns
```

```python
df.shape
```

```python
y = df['MEDV']
```

```python
y.shape
```

```python
y
```

```python
X = df[['CRIM', 'ZN', 'INDUS', 'CHAS', 'NX', 'RM', 'AGE', 'DIS', 'RAD', 'TAX', 'PTRATIO', 'B', 'LSTAT']]
```

```python
x = df.drop ('MEDV', axis=1)
```

```python
x.shape
```

```python
x
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
mm = MinMaxScaler()
```

```python
X = mm.fit_transform(X)
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.1, random_state=2529)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train, y_train)
```

```python
y_pred = lr.predict(X_test)
```

```python
y_pred.shape
```

```python
y_pred
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

```python
 mean_squared_error(y_test, y_pred)
```

```python
 mean_absolute_error(y_test, y_pred)
```

```python
 r2_score(y_test, y_pred)
```

```python
import matplotlib.pyplot as plt

plt.scatter(y_test, y_pred)
plt.xlabel("Actual Prices")
plt.ylabel("Predicted Prices")
plt.title("Actual Price vs Predicted Price")
plt.show()
```
